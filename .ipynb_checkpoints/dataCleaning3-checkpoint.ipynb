```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('./data/CSV/data.csv')
```

```python
columns_of_interest = ['Meter']

for column in columns_of_interest:
    unique_values = df[column].unique()[:10]
    print(f"{column}: {unique_values}")
```

```python
import re

numeric_pattern = re.compile(r'\b\d+\b')

def convert_to_int(value):
    match = numeric_pattern.search(str(value))
    if match:
        numeric_value = int(match.group())
        return numeric_value * 1000 if len(str(numeric_value)) in {2, 3} else numeric_value
    return None

df['Meter'] = df['Meter'].apply(convert_to_int)
```

```python
df.to_csv('./data/CSV/data.csv', index=False)
```

```python
df.dtypes
```

```python
value_counts = df['Meter'].value_counts()

print("Meter Counts:")
for value, count in zip(value_counts.index, value_counts.values):
    print(f"{value}: {count}")
```

```python
threshold_value = 500000
df['Meter'] = np.where(df['Meter'] > threshold_value, threshold_value, df['Meter'])
```

```python
df['Meter'].describe()
```

```python
df.to_csv('./data/CSV/data.csv', index=False)
```

```python
mean_value = df['Meter'].mean()

df['Meter'].fillna(mean_value.astype(int), inplace=True)
```

```python
df['Meter'] = df['Meter'].astype('Int64')
```
