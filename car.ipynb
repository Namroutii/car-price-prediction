```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def detect_outliers(data, threshold=3):
    outliers = []
    mean = np.mean(data)
    std = np.std(data)
    
    for i in data:
        z_score = (i - mean) / std 
        if np.abs(z_score) > threshold:
            outliers.append(i)
    
    return outliers
```

# ML

```python
X_cars = cars.drop('Price', axis=1)
y_cars = cars['Price']
```

```python
X_cars.shape
```

```python
y_cars.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn import metrics
import seaborn as sns
from tabulate import tabulate
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import PolynomialFeatures
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_cars,y_cars,test_size=0.2)
```

```python
ohe = OneHotEncoder()
ohe.fit(X_cars[['Car','Color','Origin']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['Car','Color','Origin']),
                                    remainder='passthrough')
```

```python
lr = LinearRegression()
knn = KNeighborsRegressor(n_neighbors=15)
dt = DecisionTreeRegressor()
rf = RandomForestRegressor()
```

```python
lr_pipe = make_pipeline(column_trans,lr)
knn_pipe = make_pipeline(column_trans,knn)
dt_pipe = make_pipeline(column_trans,dt)
rf_pipe = make_pipeline(column_trans, rf)
```

```python
y_train_1d = y_train.values.ravel()
```

```python
lr_pipe.fit(X_train,y_train)
knn_pipe.fit(X_train,y_train)
dt_pipe.fit(X_train, y_train)
rf_pipe.fit(X_train, y_train_1d)
```

```python
lr_y_pred = lr_pipe.predict(X_test)
knn_y_pred = knn_pipe.predict(X_test)
dt_y_pred = dt_pipe.predict(X_test)
rf_y_pred = rf_pipe.predict(X_test)
```

```python
lr_mse = mean_squared_error(y_test, lr_y_pred)
lr_r2 = r2_score(y_test, lr_y_pred)

knn_mse = mean_squared_error(y_test, knn_y_pred)
knn_r2 = r2_score(y_test, knn_y_pred)

dt_mse = mean_squared_error(y_test, dt_y_pred)
dt_r2 = r2_score(y_test, dt_y_pred)

rf_mse = mean_squared_error(y_test, rf_y_pred)
rf_r2 = r2_score(y_test, rf_y_pred)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_cars,y_cars,test_size=0.2)

# for PolynomialFeatures
column_trans = make_column_transformer(
    (PolynomialFeatures(degree=2, include_bias=False),['Year', 'Fuel', 'License', 'Gear',
       'Glass', 'Power', 'Meter', 'Passengers', 'Payment', 'Previous_owners',
       'Condition', 'Central_locking', 'Alarm', 'Radio', 'Sunroof',
       'Magnesium_wheels', 'Leather_seats', 'Airbag', 'For_sale',
       'For_exchange']),
    (OneHotEncoder(handle_unknown='ignore'), ['Car', 'Color', 'Origin']),
    remainder='passthrough'
)
```

```python
poly_pipe = make_pipeline(column_trans,lr)
poly_pipe.fit(X_train, y_train)
poly_y_pred = poly_pipe.predict(X_test)

poly_mse = mean_squared_error(y_test, poly_y_pred)
poly_r2 = r2_score(y_test, poly_y_pred)
```

# Polynomial Regression

```python
plt.scatter(y_test, poly_y_pred, color='black', label='Actual data')
plt.title('Actual Prices vs Predicted Prices')
plt.xlabel('Predicted Price')
plt.ylabel('Actual Prices vs Predicted Prices')
plt.legend()
plt.show()
```

# Mean Squared Error  AND  R-squared

```python
table_data = [
    ["Linear Regression", lr_mse, lr_r2],
    ["k-Nearest Neighbors", knn_mse, knn_r2],
    ["Decision Tree", dt_mse, dt_r2],
    ["Polynomial Regression", poly_mse, poly_r2],
    ["Random Forest", rf_mse, rf_r2],
]

print(tabulate(table_data, headers=["Model", "Mean Squared Error", "R-squared"]))
```

# Loop over 1000 Random States

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X_cars,y_cars,test_size=0.2,random_state=i)
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_cars,y_cars,test_size=0.2,random_state=np.argmax(scores))
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
r2_score(y_test,y_pred)
```
